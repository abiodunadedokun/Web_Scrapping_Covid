```python
# References
# 1. https://www.analyticsvidhya.com/blog/2019/10/web-scraping-hands-on-introduction-python/
# 2. https://www.dataquest.io/blog/web-scraping-python-using-beautiful-soup/
# 3. https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168#.Yh6pnOjMLIV
# 4. https://towardsdatascience.com/scraping-covid19-data-using-python-80120eb5eb66
# 5. 


from bs4 import BeautifulSoup as bs
import pandas as pd 
import requests 
import random
html_text=requests.get("https://www.worldometers.info/coronavirus")
html_text
    
```

```python
soup=bs(html_text.content, "lxml")
soup.prettify()
```

```python
# table = soup.find(id="main_table_countries_today")

# print(table)

# Search for the table and extracting it
table = soup.find('table', attrs={'id': 'main_table_countries_today'})
table
```

```python
rows=table.find_all("tr", style="")   # same as  rows = table.find_all("tr", attrs={"style":""})
data = []
for i,item in enumerate(rows):
    
    if i == 0:
        
        data.append(item.text.strip().split("\n")[1:15])
        
    else:
        data.append(item.text.strip().split("\n")[:14])
```

```python
data
```

```python

# to suppress future warnings
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import pandas as pd
import dask.dataframe as dd

dt = pd.DataFrame(data)
dt = pd.DataFrame(data[1:], columns=data[0][:14]) #Formatting the header

dt
```

```python
# Delete a column by column number
# Delete column number 4 (index number 3 in data.columns)
data = dt.drop(columns=dt.columns[0])
```

```python
data = data.drop(labels=0, axis=0)
data
```
